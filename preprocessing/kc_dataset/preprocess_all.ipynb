### 중복 처리

```python
## 노드

m = metacademy_node["name"].dropna().astype(str).str.strip()
t = tutorialbank_node["name"].dropna().astype(str).str.strip()

dup_between = sorted(set(m) & set(t))  # 두 테이블 간 중복 name
print("중복 개수:", len(dup_between))
print(*dup_between, sep="\n")
```

```python
## 엣지

def pairs(df):
    a = df["name"].dropna().astype(str).str.strip()
    b = df["prereq_name"].dropna().astype(str).str.strip()
    return set(zip(a, b))

dup_pairs = sorted(pairs(metacademy_edge) & pairs(tutorialbank_edge))
print("중복 (name, prereq_name) 개수:", len(dup_pairs))
print(*dup_pairs, sep="\n")
```
